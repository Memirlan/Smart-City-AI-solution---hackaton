```python
model = Model('vosk-model-small-ru-0.22')
```

```python
def STT(wav):
    rec = KaldiRecognizer(model, wav.getframerate())
    rec.SetWords(True)
    rec.SetPartialWords(True)

    while True:
        data = wf.readframes(4000)
        if len(data) == 0:
            break
        if rec.AcceptWaveform(data):
            continue
            # print(rec.Result())
        # else:
        #     print(rec.PartialResult())

    result = rec.FinalResult()

    return ''.join([i for i in result.split(':')[-1]]).split('"')[-2]
```

```python
wf = wave.open('33.wav', 'rb')
```

```python
STT(wf)
```

# TTS (Silero)

```python
!pip install -q torchaudio omegaconf
```

```python
import torch

language = 'en'
speaker = 'lj_16khz'
device = torch.device('cpu')
model, symbols, sample_rate, _, apply_tts = torch.hub.load(repo_or_dir='snakers4/silero-models',
                                                                      model='silero_tts',
                                                                      language=language,
                                                                      speaker=speaker)
model = model.to(device)
```

```python
from IPython.display import Audio, display
```

```python
example_text = 'Hello! My name is Jerry and I am a virtual assistant of Citix! How can I help you?'

audio = apply_tts(texts=[example_text],
                  model=model,
                  sample_rate=sample_rate,
                  symbols=symbols,
                  device=device)

display(Audio(audio[0], rate=sample_rate))
```
